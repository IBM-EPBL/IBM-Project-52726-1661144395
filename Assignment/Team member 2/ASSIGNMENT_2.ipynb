‚ö´REPLACING OUTLIERS USING MEAN IN CREDIT SCORE

```python
df.CreditScore.mean()
```

```python
q1=df.CreditScore.quantile(0.25)  #(Q1)
q3=df.CreditScore.quantile(0.75)  #(Q3)
IQR=q3-q1
lower_limit= q1 - 1.5*IQR
```

```python
df['CreditScore']=np.where(df['CreditScore']<lower_limit,650.528800,df['CreditScore'])
```
